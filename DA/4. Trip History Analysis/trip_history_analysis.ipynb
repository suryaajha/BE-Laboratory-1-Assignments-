### Doing one hot Encoding on
1. Start Station
2. End Station

```python
encoding = pd.get_dummies(df[['Start station','End station']])
df = df.drop('Start station',axis=1)
df = df.drop('End station', axis=1)
```

```python
df = pd.concat([df, encoding], axis=1)
```

```python
df.head()
```

### Label Encoding 
1. Member Type
2. Bike Number

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(df['Member type'])

df['Member type'] = le.transform(df['Member type'])

print(df['Member type'])
```

```python
le.fit(df['Bike number'])

df['Bike number'] = le.transform(df['Bike number'])

print(df['Bike number'])
```
